```python
import numpy as np
from scipy.stats import norm
import pandas as pd
import seaborn as sns
import matplotlib.pylab as plt
```

```python
nolos = pd.read_csv("./quasars_no_los.csv")
withlos = pd.read_csv("./quasars_with_los.csv")
print(len(nolos["H0"].values), len(withlos["H0"].values))

nolos = nolos[(nolos["H0"] != 0.6) & (nolos["H0"] != 0.8)]
withlos = withlos[(withlos["H0"] != 0.6) & (withlos["H0"] != 0.8)]
print(len(nolos["H0"].values), len(withlos["H0"].values))
```

```python
sns.distplot(
    nolos['H0'],
    #bins=15,
    norm_hist=True,
    rug=False, hist=False, kde=True,
    label='no-l.o.s.',
)
sns.distplot(
    withlos['H0'],
    #bins=15,
    norm_hist=True,
    rug=False, hist=False, kde=True,
    label='with-l.o.s.',
)
ax = sns.distplot(
    pd.concat([nolos['H0'], withlos['H0']], axis=0),
    #bins=15,
    norm_hist=True,
    rug=False, hist=False, kde=True,
    label='total',
)
"""
data_x, data_y = ax.lines[0].get_data()
print('data_y', np.argmax(data_y), data_x[np.argmax(data_y)])
print(np.interp(data_y.max(), data_x, data_y))
ax.plot(
    [data_x[np.argmax(data_y)], data_x[np.argmax(data_y)]],
    [0, 10],
    c='k', linestyle='--',
)
ax.text(
    data_x[np.argmax(data_y)], 1,
    '%f' % data_x[np.argmax(data_y)],
)
"""

#plt.ylim(0, 10)
plt.legend(loc='best')
```

```python
sns.distplot(nolos['MS'], rug=False, hist=False, label='no-l.o.s.')
sns.distplot(withlos['MS'], rug=False, hist=False, label='with-l.o.s.')
sns.distplot(pd.concat([nolos['MS'], withlos['MS']], axis=0), rug=False, hist=False, label='total')
plt.legend(loc='best')
```
